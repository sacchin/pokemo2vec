```python
from gensim.models import word2vec
import logging
import sys

model = word2vec.Word2Vec.load("sample.model")

def neighbor_word(posi, nega=[], n=10):
    count = 1
    result = model.most_similar(positive = posi, negative = nega, topn = n)
    for r in result:
        print(str(count)+" "+str(r[0])+" "+str(r[1]))
        count += 1

def calc(equation):
    if "+" not in equation and "-" not in equation:
        neighbor_word([equation])
    else:
        positives,negatives = [],[]
        nega_candidates = equation.split("-")
        
        for candidate in nega_candidates[0].split("+"):
            positives.append(candidate)
        
        for candidate in nega_candidates[1:]:
            if "+" in candidate:
                tmp = candidate.split("+")
                negatives.append(tmp[0])
                positives = positives + tmp[1:]
            else:
                negatives.append(candidate)
        neighbor_word(posi = positives, nega = negatives)
```
